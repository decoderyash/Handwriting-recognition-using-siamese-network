```python
import numpy as np
import pandas as pd
import keras
from keras.models import Model
from keras.layers import Input, Conv2D, MaxPooling2D, BatchNormalization, Activation, Reshape, UpSampling2D, Flatten, Dense, Dropout
from keras.models import Sequential
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
import os
import cv2
from sklearn.metrics.pairwise import cosine_similarity
```

```python
input_img = Input(shape=(64, 64, 1))
x = Conv2D(16, (3, 3), activation='relu', padding='same')(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(8, (3, 3), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(8, (3, 3), activation='relu', padding='same')(x)
encoded = MaxPooling2D((2, 2), padding='same')(x)
encoded = Flatten()(x)
encoded = Dense(8*8*8, activation='relu', name='latent')(encoded)
r = Reshape(target_shape=(8,8,8))(encoded)
x = Conv2D(8, (3, 3), activation='relu', padding='same')(r)
x = UpSampling2D((2, 2))(x)
x = Conv2D(8, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
x = Conv2D(16, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same',name='output')(x)

autoencoder = Model(input_img, decoded)
autoencoder.compile(optimizer='adadelta', loss='binary_crossentropy', metrics = ['accuracy'])
autoencoder.summary()
```

```python
path = "unseen-dataset\\TrainingSet"
folder = os.fsencode(path)
df = pd.read_csv(r"15features.csv")
l = df.columns
for i in range(1, len(l)):
    df[l[i]] = df[l[i]].astype(int) - 1
x = []
y1 = []
y2 = []
y3 = []
y4 = []
y5 = []
y6 = []
y7 = []
y8 = []
y9 = []
y10 = []
y11 = []
y12 = []
y13 = []
y14 = []
y15 = []
filenames = []
for file in os.listdir(folder):
    filename = os.fsdecode(file)
    loaded_image = cv2.imread(os.path.join("unseen-dataset\\TrainingSet",filename),0)
    print(filename)
    if len((df[df['imagename'] == filename]['letter_spacing']).tolist()) == 1:
        x.append(255.0-loaded_image.reshape((64,64,1)))
        y1.append((df[df['imagename'] == filename]['pen_pressure']).tolist()[0])
        y2.append((df[df['imagename'] == filename]['letter_spacing']).tolist()[0])
        y3.append((df[df['imagename'] == filename]['size']).tolist()[0])
        y4.append((df[df['imagename'] == filename]['dimension']).tolist()[0])
        y5.append((df[df['imagename'] == filename]['is_lowercase']).tolist()[0])
        y6.append((df[df['imagename'] == filename]['is_continuous']).tolist()[0])
        y7.append((df[df['imagename'] == filename]['slantness']).tolist()[0])
        y8.append((df[df['imagename'] == filename]['tilt']).tolist()[0])
        y9.append((df[df['imagename'] == filename]['entry_stroke_a']).tolist()[0])
        y10.append((df[df['imagename'] == filename]['staff_of_a']).tolist()[0])
        y11.append((df[df['imagename'] == filename]['formation_n']).tolist()[0])
        y12.append((df[df['imagename'] == filename]['staff_of_d']).tolist()[0])
        y13.append((df[df['imagename'] == filename]['exit_stroke_d']).tolist()[0])
        y14.append((df[df['imagename'] == filename]['word_formation']).tolist()[0])
        y15.append((df[df['imagename'] == filename]['constancy']).tolist()[0])
        filenames.append(filename)
    else:
        continue
    #print(y)
    #x.append(loaded_image) 
    #print(len(y))
train_x = np.array(x)/255.0
```
